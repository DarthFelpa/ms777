```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode, solve_ivp
```

# Modelo SIR

```python
def sir_ode(t, init, params):
    
    p, n, ν = params
    β = p*n
    
    S, I, R = init
    N = S+I+R
    
    # ODEs
    
    dS = -β*I*S
    dI = β*I*S - ν*I
    dR = ν*I
    
    return [dS, dI, dR]

def calculate_sir_r0(params):
    
    p, n, ν = params
    β = p*n
    
    return β/ν
```

```python
p = 0.05
n = 16
ν = 1/3.2

params = [p, n, ν]
init = [0.99,0.01,0]
ndays = 80

r0 = "{:.02f}".format(calculate_sir_r0(params))

print(f'Com a probabilidade de infecção em {params[0]*100}% e com os indivíduos encontrando em média outros {params[1]}, o R0 é {r0}')

sir_sol = solve_ivp(fun=lambda t, y: sir_ode(t, y, params), t_span=[0, ndays], y0=init, t_eval=np.arange(0, ndays, 1))
sir_out = pd.DataFrame({"t":sir_sol["t"],
                        "S":sir_sol["y"][0],
                        "I":sir_sol["y"][1], 
                        "R":sir_sol["y"][2],

                       })

sir_out.columns=["t", "Susceptible", "Infected", "Recovered"]

plt.style.use("ggplot")
plt.figure(figsize=(8,5))
sline = plt.plot("t","Susceptible","",data=sir_out,color="green",linewidth=2)
iline = plt.plot("t","Infected","",data=sir_out,color="orange",linewidth=2)
rline = plt.plot("t","Recovered","",data=sir_out,color="blue",linewidth=2)
plt.xlabel("Days",fontweight="bold")
plt.ylabel("Population %",fontweight="bold")
legend = plt.legend(title="",loc=5,bbox_to_anchor=(1.25,0.5))
frame = legend.get_frame()
frame.set_facecolor("white")
frame.set_linewidth(0)
```

# MODELO SEIR

```python
def seir_ode(t, init, params):
    
    p, n, μ, σ, γ  = params
    β = p*n
    
    S, E, I, R = init
    N = S+E+I+R
    
    # ODEs
    
    dS = (1-S)*μ - (β*I*S)/N
    dE = (β*I*S)/N - (σ + μ)*E
    dI = σ*E - (γ+μ)*I
    dR = γ*I - μ*R
    
    return [dS, dE, dI, dR]

def calculate_seir_r0(params):
    
    p, n, μ, σ, γ  = params
    
    β = p*n
    
    return (σ/(σ+μ))*(β/(γ+μ))
```

# Modelo SEIRD

```python
def seird_ode(t, init, params):
    
    p, n, α, σ, γ, λ  = params
    β = p*n
    
    S, E, I, R, D = init
    N = S+E+I+R
    
    # ODEs
    
    dS = - (β*I*S)/N
    dE = (β*I*S)/N - σ*E
    dI = σ*E - (1-α)*γ*I  - λ*α*I
    dR = (1-α)*γ*I 
    dD = λ*α*I
    
    return [dS, dE, dI, dR, dD]

def calculate_r0(params):
    
    p, n, α, σ, γ, λ = params
    
    β = p*n
    
    return (β/γ)
```

```python
params = [0.05, 20, 0.02, 0.25, 0.143, 0.125]
init = [0.99,0.01,0,0,0]
ndays = 80

r0 = "{:.02f}".format(calculate_r0(params))

print(f'Com a probabilidade de infecção em {params[0]*100}% e com os indivíduos encontrando em média outros {params[1]}, o R0 é {r0}')

sir_sol = solve_ivp(fun=lambda t, y: seird_ode(t, y, params), t_span=[0, ndays], y0=init, t_eval=np.arange(0, ndays, 1))
sir_out = pd.DataFrame({"t":sir_sol["t"],
                        "S":sir_sol["y"][0],
                        "E":sir_sol["y"][1],
                        "I":sir_sol["y"][2], 
                        "R":sir_sol["y"][3],
                        "D":sir_sol["y"][4]
                       })

sir_out.columns=["t", "Susceptible", "Exposed", "Infected", "Recovered", "Deceased"]

plt.style.use("ggplot")
plt.figure(figsize=(8,5))
sline = plt.plot("t","Susceptible","",data=sir_out,color="green",linewidth=2)
eline = plt.plot("t","Exposed","",data=sir_out,color="yellow",linewidth=2)
iline = plt.plot("t","Infected","",data=sir_out,color="orange",linewidth=2)
rline = plt.plot("t","Recovered","",data=sir_out,color="blue",linewidth=2)
dline = plt.plot("t","Deceased","",data=sir_out,color="black",linewidth=2)
plt.xlabel("Days",fontweight="bold")
plt.ylabel("Population %",fontweight="bold")
legend = plt.legend(title="",loc=5,bbox_to_anchor=(1.25,0.5))
frame = legend.get_frame()
frame.set_facecolor("white")
frame.set_linewidth(0)
```

# Nosso SEIRD

```python
def seird_ode(t, init, params):
    
    p, n, α, ξ, λ, σ, γ  = params
    β = p*n
    
    S, E, I, R, D = init
    N = S + E + I + R
    
    # ODEs
    
    dS = -(β*I*S)/N 
    dE = (β*I*S)/N - σ*E
    dI = σ*E - (1-α)*γ*I - α*λ*I
    dR = (1-α)*γ*I  + α*(1-ξ)*λ*I
    dD = α*ξ*λ*I
    
    return [dS, dE, dI, dR, dD]

def calculate_r0(params):
    
    p, n, α, ξ, λ, σ, γ  = params
    
    β = p*n
    
    return β/γ
```
